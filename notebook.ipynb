```python
import pandas as pd

dataset_url = "https://raw.githubusercontent.com/cogxen/datasets/main/for-laboratory/exploring-data-with-python/row-indexes/vehicles.csv"

vehicles = pd.read_csv(dataset_url)
vehicles.head()
```

```python
# Sort vehicles by best_mpge from highest to lowest. From the result, select the first 25 rows and
sort_by_fuel = vehicles.sort_values(by='fuel', ascending=True)
sort_by_fuel
```

```python
sort_by_mgpe = vehicles.sort_values(by='best_mpge', ascending=False)
result = sort_by_mgpe.iloc[:25].loc[:, 'year':'vehicle_type']
result
```

```python
# Sort vehicles by fuel from A to Z, and then for each fuel by best,mpge from highest to lowest
sort_by_fuel_best_mpge = vehicles.sort_values(by=['fuel', 'best_mpge'], ascending=[True, False])

# Show output
sort_by_fuel_best_mpge
```

```python
# Boolean Mask
is_manual = vehicles['transmission'] =='manual'

manual_vehicles = vehicles[is_manual]
manual_vehicles.head()
```

```python
# We've created a count of manual transmission vehicles by year. Add code to sort the result by year from earliest to latest.
count_by_year = manual_vehicles.groupby('year').size().sort_index()
count_by_year
```

```python
# Boolean Mask (OR)
is_suv = vehicles['vehicle_type'] == 'suv'
is_pickup = vehicles['vehicle_type'] == 'pickup'

suv_or_pickup_vehicles = vehicles[is_suv | is_pickup]
suv_or_pickup_vehicles.head()
```

```python
# Create a Boolean mask that is true when vehicle_type is neither suv nor pickup. Filter vehicles down to only the True rows.
# Boolean Mask two NOT
not_suv_pickup = ~(vehicles['vehicle_type'] == 'suv') & ~(vehicles['vehicle_type'] == 'pickup')
vehicles_not_suv_pickup = vehicles[not_suv_pickup]

vehicles_not_suv_pickup
```

```python
# Boolean Mask
is_hybrid_electric = vehicles['fuel'] == 'hybrid electric'
is_below_average = vehicles['best_mpge'] < 40

# Combined
vehicles_hybrid_below_average = vehicles[is_hybrid_electric & is_below_average]

# Show output
vehicles_hybrid_below_average
```
